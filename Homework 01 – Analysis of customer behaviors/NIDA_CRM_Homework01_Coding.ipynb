# การแสดงผลข้อมูลด้วย Matplotlib

## 1. การแสดงผลในรูปแบบ Heatmap ในความสนใจของกลุ่มข้อมูลที่เกี่ยวข้องกับอาหาร

### การเตรียมข้อมูลในส่วนของความสนใจในเรื่องของอาหาร

#### เลือก Column ที่ต้องนำมาใช้ในการแสดงผล และทำการปรับโครงสร้างเพื่อให้สามารถนำไปแสดงผลได้อย่างถูกต้อง

```python
data_bag = []
for col_index in range(1, 10):
    for answer_item in df[list(df.columns)[col_index]].tolist():
        answer_dict = {'item': list(df.columns)[col_index], 
                           'answer': answer_item, 
                           'count': 1}
       
        #print(answer_series)
        data_bag.append(answer_dict)

food_data = pd.DataFrame(data=data_bag ,columns=['item', 'answer', 'count'])
food_data
```

#### ทำการเปลี่ยนแปลงข้อมูลให้เป็นภาษาอังกฤษ เพื่อความสะดวกในการแสดงผล

```python
food_data['answer'] = food_data['answer'].replace(['ไม่สนใจอย่างมากที่สุด'], 'very very not interested')
food_data['answer'] = food_data['answer'].replace(['ไม่สนใจอย่างมาก'], 'very not interested')
food_data['answer'] = food_data['answer'].replace(['ไม่สนใจ'], 'not interested')
food_data['answer'] = food_data['answer'].replace(['เฉยๆ'], 'quietly')
food_data['answer'] = food_data['answer'].replace(['สนใจ'], 'interested')
food_data['answer'] = food_data['answer'].replace(['สนใจอย่างมาก'], 'very interested')
food_data['answer'] = food_data['answer'].replace(['สนใจอย่างมากที่สุด'], 'very very interested')
food_data
```

#### ทำการ Grouping ข้อมูล

```python
content = food_data.groupby(['item', 'answer']).sum().reset_index()
content
```

#### ทำการ Pivot Table

```python
import seaborn as sns
content = content.pivot("item", "answer" , "count")
content
```

#### ทำการจัดเรียง Column ให้เหมาะสม

```python
content = content[['very very not interested',
         'very not interested',
         'not interested',
         'quietly',
         'interested', 
         'very interested',
         'very very interested']]
content
```

### ทำการวาดกราฟ Heatmap กลุ่มความสนใจเรื่องอาหาร

```python
ax = sns.heatmap(content, annot=True, cmap='Greens')
```

### ***ผลสรุป***  
พบว่าเพื่อนร่วมชั้นชอบอาหารญี่ปุ่นเป็นอย่างมาก  
และจะเห็นได้อีกว่าชอบการรับประทานอาหารแบบ Buffet

### การเตรียมข้อมูลในส่วนของความถี่การบริโภคอาหารประเภทต่างๆ

#### เลือก Column ในส่วนของความถี่การกินอาหารต่างๆ

```python
data_bag = []
for col_index in range(31, 40):
    for answer_item in df[list(df.columns)[col_index]].tolist():
        answer_dict = {'item': list(df.columns)[col_index], 
                           'answer': answer_item, 
                           'count': 1}
       
        #print(answer_series)
        data_bag.append(answer_dict)

freq_food = pd.DataFrame(data=data_bag ,columns=['item', 'answer', 'count'])
freq_food
```
